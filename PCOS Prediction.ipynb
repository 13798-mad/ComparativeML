```python
data = data.drop(data[data["PCOS"] == "No, Yes, not diagnosed by a doctor"].index)
```

```python
sns.countplot(x="PCOS",data=data,color="purple")
plt.show()
```

```python
object_column = data.select_dtypes(include=['object']).columns
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for column in object_column:
    data[column]=le.fit_transform(data[column])
```

```python
plt.figure(figsize=(28,18))
sns.heatmap(data.corr(),annot=True)
```

```python
print(data.columns)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import r2_score,recall_score,precision_score,accuracy_score,classification_report
from sklearn.model_selection import train_test_split
```

```python
ss=StandardScaler()
features=['Age','Weight_kg', 'Menstrual_Irregularity', 'Hormonal_Imbalance',
       'Hyperandrogenism', 'Hirsutism', 'Mental_Health',
       'Conception_Difficulty', 'Insulin_Resistance', 'Diabetes',
       'Childhood_Trauma', 'Cardiovascular_Disease','Stress_Level']
x=data[features]
y=data["PCOS"]
x=ss.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
classifier={
    "Random Forest Classifier":RandomForestClassifier(),
    "Ada Boost Classifier":AdaBoostClassifier(),
    "Decision Tree Classifier":DecisionTreeClassifier()
}
```

```python
for name,clf in classifier.items():
    clf.fit(x_train,y_train)
    y_pred=clf.predict(x_test)
    accuracy=accuracy_score(y_test,y_pred)
    print(f"{name}")
    print(f"{accuracy*100}")
    print(classification_report(y_test,y_pred))
```
